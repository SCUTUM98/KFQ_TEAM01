```python
import folium
from sqlalchemy import create_engine
import pandas as pd
import time
import folium.plugins as plug
import time

total_duration = 600  # 총 작동 시간 (초)
interval = 10       # 각 작동 간격 (초)
repeat_count = 1    # 각 간격마다 반복 횟수

start_time = time.time()

while time.time() - start_time < total_duration:
    for _ in range(repeat_count):

        # SQLAlchemy 연결 설정
        engine = create_engine("mysql+mysqldb://root:tiger@localhost/db_test")


        # 쿼리 실행 및 결과를 데이터프레임으로 가져오기
        query = """
        SELECT ci.location, ce.event_type
        FROM db_test.cctv_info ci
        JOIN (SELECT cctv_id, event_type
        FROM (SELECT cctv_id, event_type 
        FROM db_test.cctv_events
        ORDER BY STR_TO_DATE(event_time, '%%Y-%%m-%%d %%H:%%i:%%s') DESC LIMIT 10) recent_events) ce ON ci.cctv_id = ce.cctv_id;
        """
        cctv_data = pd.read_sql_query(query, engine)

        # P01 = '사람 출현'
        # B01 = '주행 중인 자전거 출현'
        # B02 = '주행 중인 오토바이 출현'
        # F01 = '화재 발생'
        # F02 = '연기 탐지'
        # F03 = '차량 화재 발생'
        # A01 = '개/고양이 출몰'
        # A02 = '고라니 출몰'
        # A03 = '너구리 출몰'
        # A04 = '맷돼지 출몰'
        # S01 = '나무 감지'
        # S02 = '돌 감지'
        # S03 = '박스 감지'
        # S04 = '타이어 감지'
        # S05 = '드럼통 감지'

        event_error=['사람 출현','주행 중인 자전거 출현','주행 중인 오토바이 출현','화재 발생','연기 탐지','차량 화재 발생','개/고양이 출몰','고라니 출몰','너구리 출몰','맷돼지 출몰','나무 감지','돌 감지','박스 감지','타이어 감지','드럼통 감지']

        output_sentense=[]

        for i in range(len(cctv_data.location)):

            loc=cctv_data.location[i]
            event_type=cctv_data.event_type[i]

            if event_type == 'P01':
                event_description = event_error[0]
            elif event_type == 'B01':
                event_description = event_error[1]
            elif event_type == 'B02':
                event_description = event_error[2]
            elif event_type == 'F01':
                event_description = event_error[3]
            elif event_type == 'F02':
                event_description = event_error[4]
            elif event_type == 'F03':
                event_description = event_error[5]
            elif event_type == 'A01':
                event_description = event_error[6]
            elif event_type == 'A02':
                event_description = event_error[7]
            elif event_type == 'A03':
                event_description = event_error[8]
            elif event_type == 'A04':
                event_description = event_error[9]
            elif event_type == 'S01':
                event_description = event_error[10]
            elif event_type == 'S02':
                event_description = event_error[11]
            elif event_type == 'S03':
                event_description = event_error[12]
            elif event_type == 'S04':
                event_description = event_error[13]
            elif event_type == 'S05':
                event_description = event_error[14]
            else:
                event_description = '알 수 없는 이벤트'

            output=f"{loc}에서 {event_description}"
            output_sentense.append(output)


        # HTML 예시 코드 생성
        html_code = f"""
        <style>
            li {{
                list-style: none;
                margin-bottom: 20px;
            }}
        </style>
        <li>{output_sentense[0]}</li>
        <li>{output_sentense[1]}</li>
        <li>{output_sentense[2]}</li>
        <li>{output_sentense[3]}</li>
        <li>{output_sentense[4]}</li>
        <li>{output_sentense[5]}</li>
        <li>{output_sentense[6]}</li>
        <li>{output_sentense[7]}</li>
        <li>{output_sentense[8]}</li>
        <li>{output_sentense[9]}</li>
        """
        # HTML 파일 저장
        with open('events_list.html', 'w') as f:
            f.write(html_code)

        print("반복 작업 중...")
        time.sleep(interval)
```
