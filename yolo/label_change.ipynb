```python
import glob
import os
import shutil
```

```python
def modify_txt_file(file_path, new_value):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    modified_lines = []
    for line in lines:
        data = line.strip().split()
        if data:
            data[0] = str(new_value)
            modified_line = ' '.join(data) + '\n'
            modified_lines.append(modified_line)

    with open(file_path, 'w') as f:
        f.writelines(modified_lines)
```

```python
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
```

```python
folder_path = '/Users/bongeungu/Downloads/content 3/yolov5/Cats_Find-3/valid/labels/'
label_files = glob.glob(os.path.join(folder_path, '*.txt'))
```

```python
for i in range(len(label_files)):
    file_path = f'{label_files[i]}'
    new_value = 1
    
    modify_txt_file(file_path, new_value)
```

```python
type_list = ['images', 'labels']
for i in range(len(data_list)):
    for j in range(len(file_list)):
        original_path = f'/Users/bongeungu/Desktop/data/{file_list[j]}/{data_list[i]}/'
        print(original_path)
        
        dest_path = f'/Users/bongeungu/Desktop/data/animals/{data_list[i]}/'
        print(dest_path)
        
        for h in range(len(type_list)):
            if type_list[h] == 'images':
                original_files = glob.glob(os.path.join(original_path, 'images', '*.jpg'))
            else:
                original_files = glob.glob(os.path.join(original_path, 'labels', '*.txt'))
            print(original_files)
            
            # 대상 경로가 디렉토리인지 확인하고, 아니라면 디렉토리 생성
            if not os.path.isdir(dest_path):
                os.makedirs(dest_path)
            
            for k in range(len(original_files)):
                # 파일인지 디렉토리인지 확인
                if os.path.isfile(original_files[k]):
                    # 파일일 경우 복사
                    dest_dir = os.path.join(dest_path, type_list[h])
                    if not os.path.isdir(dest_dir):
                        os.makedirs(dest_dir)
                    shutil.copyfile(original_files[k], os.path.join(dest_dir, os.path.basename(original_files[k])))
                else:
                    print(f"Skipping directory: {original_files[k]}")
```
